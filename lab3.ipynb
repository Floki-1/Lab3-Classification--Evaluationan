### Task 1: Data Loading & Preprocessing

For this lab, we will use the Breast Cancer Wisconsin dataset. It is a binary classification problem where the goal is to predict whether a tumor is malignant or benign based on numerical measurements.


**Requirement 1:** Load the dataset from `sklearn.datasets`.


```python
# Requirement 1: Load the dataset from sklearn.datasets
# Write your code here
from sklearn.datasets import load_breast_cancer
data= load_breast_cancer()
print(data.keys()) 
```

**Requirement 2:** Split the data into a training set (80%) and a test set (20%) using `random_state=42`.


```python
# Requirement 2: Split the data into train/test sets
# Write your code here
from sklearn.model_selection import train_test_split
x=data.data
y=data.target
## this one line of code below was disscused with ai, "what is random state "?
# random state is a parameter that controls the randomness of the data splitting process. By setting a specific random state value,
#  you can ensure that the same split of data is generated each time you run the code,
#  which is useful for reproducibility and debugging purposes.
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

print("rows, columns ", x_train.shape)
print("rows, columns ", x_test.shape) 

#(rows, columns)
#455 rows means that there are 455 samples in the training set, and 30 columns means that there are 30 features for each sample.
# 114 rows means that there are 114 samples in the test set, and 30 columns means that there are 30 features for each sample.
```

**Requirement 3:** Scale the features using `StandardScaler`.

Remember the golden rule: fit the scaler on the training data only, then transform both the training and test data. Explain these parts in your code, too.


```python
# Requirement 3: Scale the features with StandardScaler
# Remember: fit on training data only, then transform train and test
# Write your code here
from sklearn.preprocessing import StandardScaler
# we import the scaler method and made an instance of it
scaler = StandardScaler()
# we fit the scaler to the training data and then transform both the training and test data using the fitted scaler.
# these 3 lines of code below were discussed with ai, "why do we fit on training data only? "
# we fit on training data only because we want to learn the parameters (mean and standard deviation)
# fit() method learns mean and standard deviation from the training data and then we use those parameters to transform both the training and test data.

scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)

```

**Requirement 4:** Verify the scaling worked. The mean should be approximately 0 and the standard deviation should be approximately 1 for the first feature in the scaled training data.


```python
# Requirement 4: Verify scaling for the first feature
# Write your code here
 
print("Mean of the first feature in the training set after scaling: ", x_train_scaled[:, 0].mean())
print("Standard deviation of the first feature in the training set after scaling: ", x_train_scaled[:, 0].std())

```

### Task 2: Logistic Regression & Model Evaluation

Accuracy is often not enough, especially in medical datasets. We need to see exactly where the model is making mistakes.


**Requirement 1:** Train a `LogisticRegression` model on your scaled training data.


```python
# Requirement 1: Train a LogisticRegression model
# Write your code here
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x_train_scaled, y_train)
```

**Requirement 2:** Make predictions on your test set.


```python
# Requirement 2: Make predictions on the test set
# Write your code here
# we use the predict() method of the trained model to make predictions on the scaled test data (x_test_scaled).
# the output 0 and 1 represent the predicted class labels for each sample in the test set, where 0 typically represents the negative class and 1 represents the positive class.
predictions = model.predict(x_test_scaled)
print(predictions)
```

**Requirement 3:** Print the `confusion_matrix` and the `classification_report`, and briefly explain the results.


```python
# Requirement 3: Print the confusion matrix and classification report
# Briefly explain the results in a comment or Markdown cell
# Write your code here
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, predictions)

print("Confusion Matrix:", cm)

print("\nClassification Report:")
print(classification_report(y_test, predictions))

# The model performed well with high accuracy and strong precision/recall values.
# The confusion matrix shows only a few incorrect predictions.
# There are very few false negatives, which is important in medical diagnosis
# because missing a cancer case can be serious. 
```
